```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/Alaska_HCU_SoilTemp_VWC_15cm.csv")

print(df.head())
print(df.info())
```

```python
df['start_time'] = pd.to_datetime(df['start_time'])
df.replace(-9999.0, pd.NA, inplace=True)
df = df[['start_time', 'latitude', 'longitude', 'soil_temp_15cm_01', 'soil_temp_15cm_02', 'vwc_15cm_cali_01', 'vwc_15cm_cali_02']]
df.columns = ['time', 'latitude', 'longitude', 'soil_temp_1', 'soil_temp_2', 'vwc_cali_1', 'vwc_cali_2']
df.set_index('time', inplace=True)
daily_df = df.resample('D').mean().reset_index()
```

```python
soil_temp_stats = daily_df[['soil_temp_1', 'soil_temp_2']].describe().T
soil_temp_stats['median'] = daily_df[['soil_temp_1', 'soil_temp_2']].median()
soil_temp_stats['std_dev'] = daily_df[['soil_temp_1', 'soil_temp_2']].std()

vwc_stats = daily_df[['vwc_cali_1', 'vwc_cali_2']].describe().T
vwc_stats['median'] = daily_df[['vwc_cali_1', 'vwc_cali_2']].median()
vwc_stats['std_dev'] = daily_df[['vwc_cali_1', 'vwc_cali_2']].std()

print("Soil Temperature Statistics (15cm):")
print(soil_temp_stats)

print("\nVolumetric Water Content (VWC) Statistics (15cm):")
print(vwc_stats)
plt.figure(figsize=(12, 6))
plt.plot(daily_df['time'], daily_df['soil_temp_1'], label='Soil Temp 1 (15cm)', color='orange')
plt.plot(daily_df['time'], daily_df['soil_temp_2'], label='Soil Temp 2 (15cm)', color='red')
plt.title("Daily Soil Temperature Over Time")
plt.xlabel("Time")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(daily_df['time'], daily_df['vwc_cali_1'], label='VWC Calibrated 1 (15cm)', color='blue')
plt.plot(daily_df['time'], daily_df['vwc_cali_2'], label='VWC Calibrated 2 (15cm)', color='green')
plt.title("Daily Volumetric Water Content Over Time")
plt.xlabel("Time")
plt.ylabel("VWC")
plt.legend()
plt.show()
```

## Conclusion

#### The soil temperature and volumetric water content (VWC) data in Alaska reflect typical Arctic climate conditions, where soil at a depth of 15 cm often remains below freezing and shows low average moisture levels, with occasional peaks likely corresponding to precipitation events or seasonal thawing. The variations in temperature and moisture suggest seasonal patterns, and these data can be crucial for analyzing the impact of climate change on polar ecosystems.
